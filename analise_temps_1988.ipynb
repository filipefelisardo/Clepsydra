```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados do Excel
file_path = 'temperaturas_1988.xlsx'
df = pd.read_excel(file_path, sheet_name='temperaturas_1988')

# Converter a coluna de datas para o formato datetime
df['DATA'] = pd.to_datetime(df['DATA'])

# Calcular a média diária para cada estação
daily_mean = df.resample('D', on='DATA').mean()

# Filtrar os dados para incluir apenas os meses de janeiro a outubro
daily_mean_filtered = daily_mean[(daily_mean.index.month >= 1) & (daily_mean.index.month <= 9)]

# Lista de estações (colunas de temperatura)
estacoes = df.columns[1:]

# Produzir os gráficos para cada estação
for estacao in estacoes:
    plt.figure(figsize=(10, 5))
    plt.plot(daily_mean_filtered.index, daily_mean_filtered[estacao], label=estacao)
    plt.title(f'Temperatura Média Diária em {estacao} - Jan a Out 1988')
    plt.xlabel('Data')
    plt.ylabel('Temperatura Média (°C)')
    plt.legend()
    plt.grid(True)
    plt.show()
```
